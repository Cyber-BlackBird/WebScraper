```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time
from selenium.webdriver.common.action_chains import ActionChains
```

```python
import time as tm
driver = webdriver.Chrome("C:/Users/Shubham/Documents/WEBDRIVER/chromedriver.exe")
driver.get("https://www.google.co.in/")
driver.maximize_window()

query = driver.find_element_by_xpath('//*[@class="gLFyf gsfi"]')
query.click()

query.send_keys("toolsvilla")
query.send_keys(Keys.ENTER)

wait = WebDriverWait(driver, 5)
driver.find_element_by_xpath('//*[@id="hdtb-msb"]/div[1]/div/div[5]/a').click()

driver.find_element_by_xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div[1]/div[2]/div/div[1]/span[1]/span/span/span[2]/span[1]/button').click()

tm.sleep(10)

action = ActionChains(driver)
for _ in range(8):
    action = action.send_keys(Keys.TAB)
    action.perform()
    
for _ in range(300):
    action.send_keys(Keys.ARROW_DOWN)
    action.perform()

users = []
review_count = []
reviews = []
response = []

rev = driver.find_elements_by_xpath('//*[@class="section-review-review-content"]')
use = driver.find_elements_by_xpath('//*[@class="section-review-title"]')
revcount = driver.find_elements_by_xpath('//*[@class="section-review-subtitle"]')
tme = driver.find_elements_by_xpath('//*[@class="section-review-publish-date"]')
resp = driver.find_elements_by_xpath('//*[@class="section-review-owner-response"]')

for r in rev :
    a = r.text
    reviews.append(a)
       
for rc in revcount :
    c = rc.text
    review_count.append(c)
    
for res in resp :
    d = res.find_element_by_class_name('section-review-text').text
    response.append(d)
    
for i in use :
    b = i.text
    users.append(b)
```
